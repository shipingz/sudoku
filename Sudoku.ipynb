```python
from random import *
import numpy as np

class Matrix:
    def __init__(self,row,col):
        self.row = row
        self.col = col
        self.matrix = np.zeros((row,col),dtype = int)
        self.aval = row * col
        
    def get(self,row,col):
        if(self.row > row and self.col > col):
            return self.matrix[(row,col)]
        else:
            return -1
        
    def set(self,row,col,item):
        if(self.row > row and self.col > col):
            self.matrix[(row,col)] = item
            self.aval = self.aval - 1;
            return 0
        else:
            return -1
    
    def printBoard(self):
        for i in range(self.row):
            for j in range(self.col):
                print(self.get(i,j), end='')
            print(' ')
        print(' ') 
    
    def good_to_add_row(self,item,index):
        for i in range(self.col):
            if self.get(i,index) == item:
                return False
        return True
    
    def good_to_add_col(self,item,index):
        for i in range(self.row):
            if self.get(index,i) == item:
                return False
        return True
    
    def good_to_add(self,item,x,y):
        r = self.good_to_add_row(item,y)
        c = self.good_to_add_col(item,x)
        return (r and c)
    

                
    
    def full_col(self,index):
        for i in range(self.row):
            if self.get(i,index) == 0:
                return False
        return True;
    
    def full_row(self,index):
        for i in range(self.col):
            if self.get(index,i) == 0:
                return False
        return True;
                
    def generate_game(self,empty,difficulty):
        # print(difficulty)
        if difficulty == 0:
            return 0
        if self.aval <= 0:
            return -1
        else:
            row = randint(0,self.row-1)
            col = randint(0,self.col-1)
            fill = randint(1,9)
            if self.get(row,col) == 0:
                res = self.set_with_rule(empty,row,col,fill)
                if res == -1:
                    self.generate_game(empty,difficulty)
                else:
                    self.generate_game(empty,difficulty-1)
            else:
                self.generate_game(empty,difficulty)
    
    def set_with_rule(self,fixed_mt,row,col,item):
        if fixed_mt.get(row,col) == 0:
            good = self.good_to_add(item,row,col)
            if good:
                # print("good")
                self.set(row,col,item)
            else:
                return -1
                
                
    def cp_board(self,source):
        for i in range(self.row):
            for j in range(self.col):
                self.set(i,j,(source.get(i,j)))
    
    
        
```

```python
empty = Matrix(9,9)
game = Matrix(9,9)
game.generate_game(empty,50)
game.printBoard()
print(game.aval)


answer = Matrix(9,9)
answer.cp_board(game)
answer.printBoard()





```

```python
answer.set_with_rule(game,0,2,8)
answer.printBoard()
answer.set_with_rule(game,0,2,1)
answer.printBoard()
```
